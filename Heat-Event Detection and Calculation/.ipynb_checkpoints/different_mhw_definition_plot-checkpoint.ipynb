### 一、2015-2024年era5的结果

```python
ds_strength_m95_2=xr.open_dataset('/home/iooleh_scf2024/sunchaofan/mhw/data/data_matlab/mhw/mhw_daily_nc/different_mhw_definition_95_2/temperature_025degree_2015_2024_m95_2_mhw_daily.nc')
ds_strength_m975_4=xr.open_dataset('/home/iooleh_scf2024/sunchaofan/mhw/data/data_matlab/mhw/mhw_daily_nc/different_mhw_definition_975_4/temperature_025degree_2015_2024_m975_4_mhw_daily.nc')
ds_strength_m90_5=xr.open_dataset('/home/iooleh_scf2024/sunchaofan/mhw/data/data_matlab/mhw/mhw_daily_nc/temperature_025degree_2015_2024_mhw_daily.nc')
mask=xr.open_dataset('/home/iooleh/ocean_grid_mask/05_ocean_grid_lonlat/grid_ocean_05_modify_360.nc')
# ds_total_event=xr.open_dataset('/home/iooleh_scf2024/sunchaofan/mhw/data/result_timelonlatcsv_dailync_countnumber/MHW_maxt2m_2015_2024_0.25_event_count.nc')
# ds_total_days=xr.open_dataset('/home/iooleh_scf2024/sunchaofan/mhw/data/result_timelonlatcsv_dailync_countnumber/MHW_maxt2m_025_2015_2024_days_count.nc')
```

```python
ds_duration_m90_5=ds_strength_m90_5.where(ds_strength_m90_5 <= 0, 1)
ds_duration_m95_2=ds_strength_m95_2.where(ds_strength_m95_2 <= 0, 1)
ds_duration_m975_4=ds_strength_m975_4.where(ds_strength_m975_4 <= 0, 1)
```

```python
inverted_mask = (1 - mask.mask).interp(lat=ds_strength_m90_5.lat, lon=ds_strength_m90_5.lon, method='nearest')
```

```python
ds_duration_m90_5_land = ds_duration_m90_5.where(inverted_mask == 1, other=np.nan)
ds_duration_m95_2_land = ds_duration_m95_2.where(inverted_mask == 1, other=np.nan)
ds_duration_m975_4_land = ds_duration_m975_4.where(inverted_mask == 1, other=np.nan)
```

```python
ds_duration_m90_5_land.t2m_mhw.isel(time=1000).plot()
```

```python
ds_duration_m90_5_land_year=((ds_duration_m90_5_land.t2m_mhw.sum(dim='time'))/10).where(inverted_mask == 1, other=np.nan)
ds_duration_m95_2_land_year=((ds_duration_m95_2_land.t2m_mhw.sum(dim='time'))/10).where(inverted_mask == 1, other=np.nan)
ds_duration_m975_4_land_year=((ds_duration_m975_4_land.t2m_mhw.sum(dim='time'))/10).where(inverted_mask == 1, other=np.nan)
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import numpy as np
import matplotlib.colors as mcolors

# 假设 ds_duration_m90_5_land_year, ds_duration_m95_2_land_year, ds_duration_m975_4_land_year 已经准备好
data_list = [ds_duration_m90_5_land_year, ds_duration_m95_2_land_year, ds_duration_m975_4_land_year]
titles = ['ERA5 2014-2025: MHW 90th, min 5 days', 'ERA5 2014-2025: MHW 95th, min 2 days', 'ERA5 2014-2025: MHW 97.5th, min 4 days']

# 设置投影
projection = ccrs.Robinson()

# 创建三张子图
fig, axes = plt.subplots(1, 3, figsize=(15, 4), subplot_kw={'projection': projection})
#plt.subplots_adjust(hspace=0.1, wspace=0.05)

plt.rcParams['font.family'] = 'Nimbus Roman'
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['axes.unicode_minus'] = True

# 设置不同的 colorbar 范围
levels_list = [
    np.arange(0, 44, 4),  # 对于第一个子图的 colorbar 范围
    np.arange(0, 55, 5),  # 对于第二个子图的 colorbar 范围
    np.arange(0, 22, 2)   # 对于第三个子图的 colorbar 范围
]

# 绘制图形
for ax, data, title, levels in zip(axes, data_list, titles, levels_list):
    c = ax.contourf(data.lon, data.lat, data, cmap='RdBu_r', levels=levels, extend='both', transform=ccrs.PlateCarree())
    ax.add_feature(cfeature.COASTLINE, color='gray')
    ax.set_title(title, fontsize=14)

    # 为每个子图单独添加 colorbar
    cbar = fig.colorbar(c, ax=ax, shrink=0.6, fraction=0.03, pad=0.01, extend='both')
    cbar.ax.set_ylabel('(days/year)', fontsize=16)
    cbar.ax.tick_params(labelsize=14)

plt.show()
```
